```python
rows_with_all_units=[]
for x in range(len(nutri_df)):
    if nutri_df.iloc[x].str.contains("[a-zA-Z]",regex =True).sum() == 72:
        rows_with_all_units.append(x)
        break
    else:
        continue
```

```python
def find_rows_with_all_unit(df,threshold=72):
    matching_rows=[]
    for row in range(len(df)):
        alpha_count = df.iloc[row].str.contains('[a-zA-Z]',regex = True).sum()
        if alpha_count==threshold:
            matching_rows.append(row)
        else:
            continue
    return matching_rows
```

```python
rows_with_all_unit= find_rows_with_all_unit(nutri_df)
```

```python
original_col=list(nutri_df.iloc[24].str.replace('[0-9 .]','',regex=True).index)
```

```python
units=list(nutri_df.iloc[24].str.replace('[0-9 .]','',regex=True).values)
```

```python
new_column=[]
for col,unit in zip(original_col,units):
    new_col=f"{col}_{unit}"
    new_column.append(new_col)

```

```python
coldict=dict(zip(original_col,new_column))
```

```python
df.rename(columns = coldict,inplace =True)
```
